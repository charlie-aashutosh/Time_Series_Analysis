```python
import pandas as pd
import datetime as datetime
import pandas_datareader as pdr
```

```python
import yfinance as yf #library to fetch data directy from Yahoo finance
```

```python
stock_data=yf.download('TSLA',start='2023-10-01',end='2024-02-17')
```

```python
stock_data.head()
```

```python
stock_data['Open'].plot(figsize=(15,5))
```

```python
#simple moving average - smoothening the curve
```

```python
stock_data['Open:10 days rolling']=stock_data['Open'].rolling(window=10,min_periods=1).mean()
```

```python
stock_data[['Open','Open:10 days rolling']].plot(figsize=(15,5))
```

```python
stock_data['Open:10 days rolling']=stock_data['Open'].rolling(window=10,min_periods=1).mean()
stock_data['Open:30 days rolling']=stock_data['Open'].rolling(window=30,min_periods=1).mean()
stock_data['Open:50 days rolling']=stock_data['Open'].rolling(window=50,min_periods=1).mean()
```

```python
stock_data[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(figsize=(15,5))
```

```python
#using expanding() - Cumulative Moving Average
stock_data['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#exponential moving average - more focus (more weights) on recent data
#smoothing factor 0.1
stock_data['EMA_0.1']=stock_data['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
stock_data[['Open','EMA_0.1']].plot(xlim=['2023-11-01','2024-02-01'],figsize=(10,5))
```

```python
stock_data['EMA_0.3']=stock_data['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
stock_data[['Open','EMA_0.1','EMA_0.3']].plot(xlim=['2023-11-01','2024-02-01'],figsize=(10,5))
```

```python
#Exponential Weighted Moving Average
stock_data['EMA_5Days']=stock_data['Open'].ewm(span=5).mean()
```

```python
stock_data[['Open','EMA_5Days']].plot()
```

```python
stock_data[['Open','EMA_0.1','EMA_5Days']].plot(xlim=['2023-11-01','2024-02-01'],figsize=(10,5))
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python

def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adf_test(data['Thousands of Passengers'])
```

```python
## Use Techniques Differencing
data['Passengers First Difference']=data['Thousands of Passengers']-data['Thousands of Passengers'].shift(1)
```

```python
data.head()
```

```python
adf_test(data['Passengers First Difference'].dropna())
```

```python
data['Passengers Second Difference']=data['Passengers First Difference']-data['Passengers First Difference'].shift(1)
```

```python
adf_test(data['Passengers Second Difference'].dropna())
```

```python
## Use Techniques Differencing for 12 months
data['Passengers 12 Difference']=data['Thousands of Passengers']-data['Thousands of Passengers'].shift(12)
```

```python
adf_test(data['Passengers 12 Difference'].dropna())
```

```python
#splitting the data
from datetime import datetime,timedelta
train_dataset_end=datetime(1955,12,1)
test_dataset_end=datetime(1960,12,1)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_SARIMA=SARIMAX(train_data['Thousands of Passengers'],order=(3,0,5),seasonal_order=(0,1,0,12))
```

```python
model_SARIMA_fit=model_SARIMA.fit()
```

```python
model_SARIMA_fit.summary()
```
